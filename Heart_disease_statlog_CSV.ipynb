```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
df = pd.read_csv("Heart_disease_statlog.csv")
```

```python
df.info()
```

```python
def convert_cp(value):
    if value == 0:
        return "Typical Angina"
    elif value == 1:
        return "Atypical Angina"
    elif value == 2:
        return "Non-Anginal Pain"
    elif value == 3:
        return "Asymptomatic"
    else:
        return "Unknown"  # सुरक्षा के लिए, अगर कोई अनजान वैल्यू हो

df["cp"] = df["cp"].apply(convert_cp)
df["cp"] = df["cp"].astype("str")
```

```python
df.describe()
```

```python
def conv(value):
    if value == 1:  # if के बाद `:` होना चाहिए
        return "Female"
    else:
        return "Male"

df["sex"] = df["sex"].apply(conv)
df["sex"] = df["sex"].astype("str")
```

```python
def convert_fbs(value):
    if value == 1:
        return "High"
    else:
        return "Normal"

df["fbs"] = df["fbs"].apply(convert_fbs)
df["fbs"] = df["fbs"].astype("str")
```

```python
def convert_restecg(value):
    if value == 0:
        return "Normal"
    elif value == 1:
        return "ST-T Wave Abnormality"
    elif value == 2:
        return "Left Ventricular Hypertrophy"
    else:
        return "Unknown" 

df["restecg"] = df["restecg"].apply(convert_restecg)
df["restecg"] = df["restecg"].astype("str")
```

```python
def convert_exang(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df["exang"] = df["exang"].apply(convert_exang)
df["exang"] = df["exang"].astype("str")
```

```python
def convert_slope(value):
    if value == 0:
        return "Up Sloping"
    elif value == 1:
        return "Flat"
    elif value == 2:
        return "Down Sloping"
    else:
        return "Unknown"  

df["slope"] = df["slope"].apply(convert_slope)
df["slope"] = df["slope"].astype("str")
```

```python
def convert_thal(value):
    if value == 1:
        return "Normal Blood Flow"
    elif value == 2:
        return "Fixed Defect (No Blood Flow)"
    elif value == 3:
        return "Reversible Defect (Abnormal Blood Flow)"
    else:
        return "Unknown"  

df["thal"] = df["thal"].apply(convert_thal)
df["thal"] = df["thal"].astype("str")
```

```python
def convert_target(value):
    if value == 1:
        return "Heart Disease"
    else:
        return "Normal"

df["target"] = df["target"].apply(convert_target)
df["target"] = df["target"].astype("str")
```

```python
df.to_csv("Heart_disease_statlog.csv")
```

```python

numrical_cols = ["age", "trestbps", "chol", "thalach", "oldpeak", "ca"]  
object_cols = ["sex", "cp","fbs","restecg","exang","slope","thal","target"]
```

# Generate Correlation Heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[numrical_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

# Generate Boxplots for Outlier Detection

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))
fig.suptitle("Boxplots for Outlier Detection", fontsize=16)
axes = axes.flatten()

for i, col in enumerate(numrical_cols):
    sns.boxplot(y=df[col], ax=axes[i])
    axes[i].set_title(col)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# सेट करें figure का साइज
plt.figure(figsize=(15, 12))

# हर categorical column के लिए bar plot बनाएँ
for i, cat_col in enumerate(object_cols, 1):
    plt.subplot(4, 2, i)  # 4x2 ग्रिड में प्लॉट्स बनाएँ
    sns.barplot(x=cat_col, y="age", data=df, ci=None, palette="coolwarm")  # उदाहरण के लिए "age" का उपयोग किया गया है
    plt.title(f"Average Age by {cat_col}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 15))

# हर numerical column के लिए plot बनाएँ
for i, num_col in enumerate(numrical_cols, 1):
    plt.subplot(3, 2, i)  # 3x2 ग्रिड में प्लॉट्स बनाएँ
    sns.barplot(x="target", y=num_col, data=df, ci=None, palette="viridis")
    plt.title(f"{num_col} Distribution by Target")

plt.tight_layout()
plt.show()
```
